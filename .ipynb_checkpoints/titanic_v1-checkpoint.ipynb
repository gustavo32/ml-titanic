```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.linear_model import LassoCV
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV, SelectKBest
sns.set()
```

```python
def create_treatment_column(single_data):
    """This function create a treatment's list:
    1 - Miss.
    2 - Mr.
    3 - Mrs.
    4 - Master
    5 - Others"""
    if re.findall(r"(.+(M|m)iss.+)|(.+(M|m)ile.+)|(.+(M|m)s.+)", single_data):
        return "Miss"
    elif re.findall(r"(.+(M|m)rs.+)|(.+(M|m)me.+)", single_data):
        return "Mrs"
    elif re.findall(r".+(M|m)r.+", single_data):
        return "Mr"
    elif re.findall(r".+(M|m)aster.+", single_data):
        return "Master"
    return "Others"  # Others


titanic_data = pd.read_csv("titanic_data.csv")
test_data = pd.read_csv("test.csv")
test_data = test_data.iloc[:, 1:]
titanic_data = titanic_data.iloc[:, 1:]
for df in [titanic_data, test_data]: 
    df["Treatment"] = df["Name"]
    df["Treatment"] = df["Treatment"].apply(create_treatment_column)
```

```python
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    #1º argumento, qual cor em decimal se inicia. 
    #2º argumento qual cor se acabará.
    #3º (as_cmap) Suavizar a cor de acordo com o número.
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, #altura da barra ao lado
        ax=ax,
        annot=True, #colocar os números em cada frame
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Correlação de Pearson através das características', y=1.05, size=15)

correlation_heatmap(titanic_data)
```

# Feature Engineering

## Creating New Features

```python
def create_new_features(df):
    # Creating Family Size
    df["FamSize"] = df["Parch"] + df["SibSp"] + 1
    df.drop(columns=["Parch", "SibSp"], inplace=True)
    # Creating is alone or not
    df["isAlone"] = df["FamSize"] == 1
    df["Cabin_null"] = df["Cabin"].isnull()
    unique_cabins = pd.unique(df["Cabin"].astype("str").str[0])
    dic = {}
    for i, c in enumerate(unique_cabins):
        dic[c] = i
    df["Cabin"] = df["Cabin"].astype("str").str[0].map(dic)
    df["Age_null"] = df["Age"].isnull()
    return df
titanic_data = create_new_features(titanic_data)
test_data = create_new_features(test_data)
titanic_data
```

## Get dummies

```python
# def dummies(df):
#     sex_map = {"male": 0, "female": 1}
#     embarked_map = {"S": 0, "C": 1, "Q": 2}
#     treatment_map = {"Mr": 0, "Mrs": 1, "Miss": 2, "Master": 3, "Others": 4}
#     df["Sex"] = df["Sex"].map(sex_map)
#     df["Embarked"] = df["Embarked"].map(embarked_map)
#     df["Treatment"] = df["Treatment"].map(treatment_map)
#     return df

def dummies(df):
    dummies_treatment = pd.get_dummies(df["Treatment"])
    dummies_embarked = pd.get_dummies(df["Embarked"])
    dummies_sex = pd.get_dummies(df["Sex"])
    df.drop(columns=["Sex", "Embarked", "Treatment"], inplace=True)
    return pd.concat([df, dummies_sex, dummies_treatment, dummies_embarked], axis=1)

titanic_data = dummies(titanic_data)
test_data = dummies(test_data)
titanic_data.iloc[:10, 2:]
```

## Fare and Age bins

```python
for df in [titanic_data, test_data]:
    bins = [-1, 20, 50, 100, 250, 10000]
    cat = [0, 1, 2, 3, 4]
    df["Fare"] = pd.cut(df["Fare"], bins, labels=cat)
    df["Fare"] = pd.to_numeric(df["Fare"])

    bins = [-1, 4, 10, 16, 25, 40, 60, 2000]
    cat = [0, 1, 2, 3, 4, 5, 6]
    df["Age"] = pd.cut(df["Age"], bins, labels=cat)
    df["Age"] = pd.to_numeric(df["Age"])
```

## Criando uma regressão pelo método Lasso com validação cruzada para prever as idades

```python
cols = ["FamSize", "isAlone", "Pclass", "Master", "Miss", "Mr", "Mrs", "Others", "C", "Q", "S", "Cabin",
        "Cabin_null", "Age"]
# cols = ["FamSize", "isAlone", "Pclass", "Treatment", "Embarked", "Cabin",
#         "Cabin_null", "SibSp", "Parch", "Age"]
dropped_nan = titanic_data[cols].dropna().values
Y = dropped_nan[:, -1]
X = dropped_nan[:, :-1]
reg = LassoCV(cv=5).fit(X, Y)
reg.score(X, Y)
```

## Criando uma regressão com Random Forest

```python
reg = RandomForestRegressor(n_estimators=100).fit(X,Y)
reg.score(X, Y)
```

```python
c = cols[:-1]
titanic_data[c].isnull().sum()
for df in [titanic_data, test_data]:
    df["Fare"].fillna(df["Fare"].mean(), inplace=True)
    df.dropna(inplace=True)
    df["Age"].where(df["Age"].notnull(), reg.predict(df[c]), inplace=True)
    df[df["Age"] < 0] = 0
```

## Criando modelos preditivos

```python
params = {'kernel' : ['rbf'], 
          'gamma': [0.001, 0.005, 0.01, 0.02, 0.05, 0.1],
          'C': [10, 30, 50, 100, 500]}

X = titanic_data.drop(columns=["Survived", "Name", "Ticket"])
Y = titanic_data["Survived"]
# print(X.shape)
# new_X = SelectKBest(k=15).fit_transform(X, Y)
# print(new_X.shape)
clf_svm = SVC(C=10, gamma=0.02)
scores = cross_val_score(clf_svm, X, Y, cv=5)
print("Acurácia Count Vector: ", round(scores.mean(), 2), " +/- (", round(scores.std(), 2), ")")

# clf_svm = GridSearchCV(clf_svm, params, cv=5)
clf_svm.fit(X, Y)
```
